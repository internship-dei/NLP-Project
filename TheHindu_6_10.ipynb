# **SCRAPING DATA FROM 6/1/21 TO 10/1/21**

```python
# Importing all the required libraries
import requests
from bs4 import BeautifulSoup 
from google.colab import files
import json
```

```python
# Function to get a soup from a given url using proxies 
def get_soup(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  return soup
```

```python
# Accessing TheHindu site
soup  = get_soup('https://www.thehindu.com/archive/web/2021/01/')
```

```python
# Getting the links to all the days
table_headers = soup.find_all('th')
table_body = table_headers[0].find_next('tbody')
url_list = []
for tr in table_body.find_all('td'):
  for a in tr.find_all('a', href=True):
    url_list.append(a['href'])
print("Links to the pages from 1/1/21 to 31/1/21: \n", url_list)
```

```python
# Getting all the links using URL Pattern recognition 
string1 = 'https://www.thehindu.com/archive/web/2021/01/'
URL_List = []
for  i in range(6, 11):
  if i < 10:
    url = string1+ f'/0{i}/'
  elif i==10:
    url = string1 + f'/{i}/'
  URL_List.append(url)
```

```python
# Finding all the links from 6/1/21 to 10/1/21
all_links =[]
all_text = []
for link in url_list[5:10]:
  temp_soup = get_soup(link)
  temp_div = temp_soup.find("div", class_="tpaper-container") 
  for lines in temp_div.find_all('li'):
    for link1 in lines.find_all('a', href= True):
      all_links.append(link1['href'])
```

```python
# Function to convert lists to strings 
def listToString(s): 
        # initialize an empty string
    str1 = "" 
        # traverse in the string  
    for ele in s: 
        str1 += ele     
    # return string  
    return str1
```

```python
# Getting all the paragraph elements in all the links scraped
p_list = []
for links in all_links:
  second_temp_soup = get_soup(links)
  data = ''
  str = []
  for data in second_temp_soup.find_all('p'):
    str.append(data.get_text())
    finalstring = listToString(str)
  p_list.append(finalstring) 
```

```python
# Length of the Paragraph-list formed 
len(p_list)
```

```python
# getting the number of doccuments in each day-URL
count_date_no = 5
for link in url_list[5:10]:
  count_date_no = count_date_no+1
  count_doc_no = 0
  temp_soup = get_soup(link)
  temp_div = temp_soup.find("div", class_="tpaper-container") 
  for lines in temp_div.find_all('li'):
    for link1 in lines.find_all('a', href= True):
      count_doc_no = count_doc_no +1
  print(f"day {count_date_no} has {count_doc_no} documents")
 
```

```python
# getting a list consisting of the names of all the documents 
doc_names= []
page_links_count = [431, 448, 388, 394, 338]
doc_name = []
index = 0
for i in range(11,15):
  for j in range(page_links_count[index]):
    filename = 'thehindu_jan_' + str(i) + '_file_' + str(j+1) + '.json'
    doc_name.append(filename)
  index += 1
doc_name 
```

```python
# Creating and Downloading all the .json files
import json
for i in range(len(doc_name)):
  dict2={}
  dict2[doc_names[i]] = p_list[i]
  with open(doc_names[i], "w") as f:
    json.dump(dict2, f)
```

```python
# Changing the location of all the files downloaded to their respective folders on the Google Drive
import shutil
index = 0
for i in range(6, 11):
  for j in range(1, page_links_count[index] + 1):
    shutil.move(f"/content/thehindu_jan_{i}_file_{j}.json", f"/content/drive/MyDrive/Jan{i}")
  index += 1
```
